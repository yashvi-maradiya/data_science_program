```python
'''Problem: Given a CSV file containing sales data with columns: 'Date', 'Product', 'Units Sold', and 'Revenue', write a Python script using Pandas to:

Read the data into a DataFrame.
Calculate the total revenue generated each day.
Plot a line graph showing the daily revenue over time.'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the data into a DataFrame.
```

```python
data = pd.read_csv("sales_data.csv")
```

```python
#Calculate the total revenue generated each day.
```

```python
total_rev = data.groupby('Date')['Revenue'].sum().reset_index()
```

```python
total_rev
```

```python
data['Date'] = pd.to_datetime(data['Date'])
daily_revenue = data.groupby('Date')['Revenue'].sum()
plt.plot(daily_revenue.index, daily_revenue.values)
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.title('Daily Revenue graph')
plt.grid(True)
plt.xticks(rotation=45)
# Save or display the plot
plt.savefig('daily_revenue.png')
plt.show()
```
