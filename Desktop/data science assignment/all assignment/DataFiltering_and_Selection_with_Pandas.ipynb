```python
'''Problem: Use Pandas to load a dataset containing information about movies, including columns: 
'Title', 'Genre', 'Year', and 'Rating'. Write a Python script to:

Filter the DataFrame to select movies released after 2010 with a rating above 8.
Display the top 10 movies based on their ratings.
Count the number of movies in each genre category.'''
```

```python
import pandas as pd
```

```python
data = pd.read_csv("movies_data.csv")
```

```python
data
```

```python
#Filter the DataFrame to select movies released after 2010 with a rating above 8.
filter_data = data[(data['Year'] > 2010) & (data['Rating'] > 8)]
filter_data
```

```python
#Display the top 10 movies based on their ratings.
top_10 = data.nlargest(10,'Rating')
top_10
```

```python
#Count the number of movies in each genre category.
count_movies = data['Genre'].value_counts().reset_index()
count_movies
```
